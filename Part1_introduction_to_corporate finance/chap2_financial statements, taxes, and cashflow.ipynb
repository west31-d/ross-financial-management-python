```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#we use yfinanace library 4 make example 
import yfinance as yf

#show big numbers with commas instead of scientific notation
pd.options.display.float_format = '{:,.0f}'.format
```

```python
#call balance_sheet 2 Apple > All examples are provided with Apple's stock data
ticker = yf.Ticker("AAPL")

#yearly
bs = ticker.balance_sheet

#quarterly
bs_quarterly = ticker.quarterly_balance_sheet

print(bs)

#check index
bs.index.tolist()
```

```python
#net working capital = current asset - current liabilities

current_asset = bs.loc["Current Assets"]
current_liabilites = bs.loc["Current Liabilities"]

net_working_capital = current_asset - current_liabilites

print(net_working_capital)
```

```python
# shareholder's equity = asset - liabilites
# asset - (equity + liabilites) = 0

total_asset = bs.loc["Total Assets"]
total_liabilites = bs.loc["Total Liabilities Net Minority Interest"]
total_equity = bs.loc["Total Equity Gross Minority Interest"]

check = total_asset - (total_equity + total_liabilites)

result = pd.DataFrame({
    "total asset": total_asset,
    "total equity": total_equity,
    "total liabilites": total_liabilites,
    "total equity + total liabilites" : total_liabilites + total_equity,
    "diff": check
    })

result
```

```python
# call income statement 4 Apple

#yearly
income_stmt = ticker.income_stmt

#quarterly
income_stmt_quarterly = ticker.quarterly_income_stmt

print(income_stmt)
income_stmt.index.tolist()
```

```python
#revenues - expenses = income

total_revenues = income_stmt.loc["Total Revenue"]
total_expenses = income_stmt.loc["Total Expenses"] + income_stmt.loc["Tax Provision"] 
    # There's no built-in "total expenses" field, so I added the tax provision manually
net_income = income_stmt.loc["Net Income"]

check = total_revenues - total_expenses

result = pd.DataFrame({'revenues': total_revenues,
                       'expenses' : total_expenses,
                       'income': net_income,
                       'diff' : check,
                       })

result
```

```python
#simple income statement

# dividends is on cashflow, not income statement so I called cashflow
cashflow = ticker.cashflow

# and addition to retained Earnings is a cumulative balance, so the increase is computed
# by subtracting the prior period's balance from the current period's.
 
retained_earnings = bs.loc["Retained Earnings"]
addition_to_retained_earnings = retained_earnings - retained_earnings.shift(-1)


simple_income_statement = pd.DataFrame({
    "net sales" : income_stmt.loc["Total Revenue"],
    "cost of goods sold" : income_stmt.loc["Cost Of Revenue"],
    "depreciation" : income_stmt.loc["Reconciled Depreciation"],
    "EBIT" : income_stmt.loc["EBIT"],
    "interest paid" : income_stmt.loc["Interest Expense"],
    "EBT" : income_stmt.loc["Pretax Income"],
    "taxes" : income_stmt.loc["Tax Provision"],
    "net income" : income_stmt.loc["Net Income"],
    "dividends" : cashflow.loc["Cash Dividends Paid"],
    "addition to retained earnings" : addition_to_retained_earnings
})

check = (simple_income_statement["net income"] 
         + simple_income_statement["dividends"]   
         - simple_income_statement["addition to retained earnings"])

simple_income_statement["check"] = check
simple_income_statement

# A small gap can remain due to share buybacks and other equity adjustments
# (e.g. stock-based compensation) that aren't fully captured here.
```

```python
#marginal tax rate calculator
tax_brackets = [
        (9875, 0.10),
        (40125, 0.12),
        (85525, 0.22),
        (163300, 0.24),
        (207350, 0.32),
        (518400, 0.35),
        (float("inf"), 0.37)
    ]

taxable_income = 10000000

def marginal_tax_rate(taxable_income):
    tax = 0
    lower = 0
    marginal_rate = None

    for upper, rate in tax_brackets:
        if taxable_income > lower:
            taxed_amount = min(taxable_income, upper) - lower
            tax += taxed_amount * rate
            marginal_rate = rate

        lower = upper
        if taxable_income <= upper:
            break

    average_rate = tax / taxable_income if taxable_income > 0 else 0
    
    return tax, marginal_rate, average_rate

tax, marginal_rate, average_rate = marginal_tax_rate(taxable_income)

print(taxable_income)
print("tax: ", tax)
print("marginal_rate: ", marginal_rate)
print("average_rate: ", average_rate)
```

```python
#Cashflow
#cash flow from assets(CFFA) = cash flow from creditors(CFTC) + cash flow to stockholders(CFTS)

#CFFA is operating cash flow(OCF) - net capital spending(NCS) - net working captial(NWC)

#OCF is EBIT + Depreciation - Taxes
ocf = income_stmt.loc["EBIT"] + income_stmt.loc["Reconciled Depreciation"] - income_stmt.loc["Tax Provision"]

#NCS is Ending Net Fixed Assets - Beginning Net Fixed Assets + Depreciation
ncs = bs.loc["Net PPE"] - bs.loc["Net PPE"].shift(-1) + income_stmt.loc["Reconciled Depreciation"]

#NWC is Ending Working Captial - Beginning Working Capital
nwc = (bs.loc["Current Assets"] - bs.loc["Current Assets"].shift(-1)) - (bs.loc["Current Liabilities"] - bs.loc["Current Liabilities"].shift(-1))

cffa = ocf - ncs - nwc

#CFTC is Interest Paid - Net New Borrowing
#Net New Borrowing is Ending Long term Debt - Beginning Long term Debt
net_new_borrowing = bs.loc["Long Term Debt"] - bs.loc["Long Term Debt"].shift(-1)

cftc = income_stmt.loc["Interest Expense"] - net_new_borrowing

#CFTS is Dividends Paid - Net New Equity Raised
#Net New Equity is Ending Common Stock - Beginning Common Stock
net_new_equity = bs.loc["Common Stock"] - bs.loc["Common Stock"].shift(-1)

cfts = -cashflow.loc["Cash Dividends Paid"] - net_new_equity

check = cftc + cfts

if cffa == check:
    print("ok")
else:
    print("no")
    print("check")
```
