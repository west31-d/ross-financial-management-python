```python
#Import packages
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
#Future value calculator
def fv(pv, r, n):
    fv = pv*(1+r) ** n
    return fv

n = np.linspace(0,10,100)
r = np.array([0,0.05,0.10,0.15,0.20])

n,r = np.meshgrid(n,r)

future_value = fv(1,r,n)

future_value
```

```python
#Plot future value growth for each interest rate /  It is made by claude code
import matplotlib.pyplot as plt

colors = ["#2a78d6", "#008300", "#e87ba4", "#eda100", "#1baf7a"]
labels = ["0%", "5%", "10%", "15%", "20%"]

fig, ax = plt.subplots(figsize=(8, 5))
for i in range(future_value.shape[0]):
    ax.plot(n[i], future_value[i], color=colors[i], linewidth=2)
    ax.annotate(labels[i], xy=(n[i, -1], future_value[i, -1]), xytext=(4, 0),
                textcoords="offset points", va="center", color="#0b0b0b", fontsize=9)

ax.set_xlabel("Number of Periods (n)", color="#52514e")
ax.set_ylabel("Future Value of $1", color="#52514e")
ax.set_title("Future Value Growth for Different Interest Rates", color="#0b0b0b")
ax.grid(True, color="#e1e0d9", linewidth=0.8)
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
for spine in ["left", "bottom"]:
    ax.spines[spine].set_color("#c3c2b7")
ax.tick_params(colors="#898781")

plt.tight_layout()
plt.show()
```

```python
#Discount Present Value Calculator
def pv(fv, r, n):
    pv = fv/((1+r)**n)
    return pv

n = np.linspace(0,10,100)
r = np.array([0,0.05,0.10,0.15,0.20])

n,r = np.meshgrid(n,r)

present_value = pv(1,r,n)

present_value
```

```python
#Plot present value discount for each interest rate /  It is made by claude code
import matplotlib.pyplot as plt

colors = ["#2a78d6", "#008300", "#e87ba4", "#eda100", "#1baf7a"]
labels = ["0%", "5%", "10%", "15%", "20%"]

fig, ax = plt.subplots(figsize=(8, 5))
for i in range(future_value.shape[0]):
    ax.plot(n[i], present_value[i], color=colors[i], linewidth=2)
    ax.annotate(labels[i], xy=(n[i, -1], present_value[i, -1]), xytext=(4, 0),
                textcoords="offset points", va="center", color="#0b0b0b", fontsize=9)

ax.set_xlabel("Number of Periods (n)", color="#52514e")
ax.set_ylabel("Present Value", color="#52514e")
ax.grid(True, color="#e1e0d9", linewidth=0.8)
for spine in ["top", "right"]:
    ax.spines[spine].set_visible(False)
for spine in ["left", "bottom"]:
    ax.spines[spine].set_color("#c3c2b7")
ax.tick_params(colors="#898781")

plt.tight_layout()
plt.show()
```

```python
#Determining the Discount Rate

def r(pv, fv, t):
    r = (fv/pv) ** (1/t) - 1
    return r

print(r(1250, 1350, 1))
```

```python
#Finding the Number of Periods

def t(pv, fv, r):
    t = np.log(fv/pv) / np.log(1+r)
    return t

print(t(25000, 50000, 0.12))
```
